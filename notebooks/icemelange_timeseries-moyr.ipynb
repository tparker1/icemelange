## Define Region of Interest
### Full-length of Jakobshavn Fjord

```python
# path to geoJSON polygon
geoJSON_filepath = '/home/jovyan/tutorial/Jakobshavn_GJSON.geojson'
region = sliderule.toregion(geoJSON_filepath)["poly"]

region
```

## Jakobshavn Fjord - Full Fjord

```python
# Build SlideRule request
# Define parameters
parms = {
    "poly": region,
    "srt": icesat2.SRT_SEA_ICE,
    "cnf": icesat2.CNF_SURFACE_HIGH,
    "ats": 5.0,
    "len": 20.0,
    "res": 20.0,
    "maxi": 5
}

# Calculated ATL06 dataframe
is2_df = icesat2.atl06p(parms, "nsidc-s3")
```

```python
#Remove outlier data
is2_df= is2_df[is2_df['h_mean'] < 175]
is2_df= is2_df[is2_df['h_mean'] > -75]

# Display top of the ICESat-2 dataframe
#is2_df.head()
```

## Plot Time Series 

```python
# Add a date column to the data frame so we can group the data by day 

# get the timestamps from the dataframe index
timestamps = is2_df.index.tolist()

# get the YYYY-MM-DD for each row
time = []
moyr = []
for t in timestamps:
    time.append(str(t)[0:10])
    moyr.append(str(t)[0:7])
    
# store YYYY-MM-DD in a new column called day
is2_df['day'] = time
is2_df['moyr'] = moyr
```

```python
#%matplotlib widget
fig, ax = plt.subplots(figsize=(20, 6))

plt.plot(is2_df.groupby(['moyr']).mean()['h_mean'])

plt.xticks(rotation=45)

plt.show()
```
